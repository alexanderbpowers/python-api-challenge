```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import gmaps
from api_keys import g_key
from api_keys import weather_api_key
import citipy
```

## Generate Cities List

```python
cities = pd.read_csv("../cities.csv")
cities_list = cities["City"].unique()
print(len(cities_list))
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
weather = []
number = 1
url = "https://api.openweathermap.org/data/2.5/weather?"
# lat = cities.loc[cities["City"]==city, "Lat"]
# lng = cities.loc[cities["City"]==city, "Lng"]

for city in cities_list:
    lat = cities.loc[cities["City"]==city, "Lat"]
    lng = cities.loc[cities["City"]==city, "Lng"]
    params = {
    'lat':lat,
    'lon':lng,
    'appid':weather_api_key
    }
    
    
    response = requests.get(url, params)
    data = response.json()
    weather = data["weather"][0]["main"]
    
    if data == "":
        print(f"{number}, {city}: NO DATA AVAILABLE")
    else:
        print(f"{number}, {city}'s weather: {weather}")
    
    number += 1

    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cities.to_csv("city_data", index=False)
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
cities_humid = cities.loc[cities["Humidity"] > 100]
cities_humid
```

```python
#  Get the indices of cities that have humidity over 100%.
humid_index= cities_humid.index
humid_index[0]
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".
clean_city_data = cities.drop(humid_index[0], inplace=False)
clean_city_data["Max Temp_C"]= round((clean_city_data["Max Temp"] - 32)/1.8, 2)
clean_city_data
```

```python
clean_city_data.to_csv("clean_city_data", index=False)
```

## Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

## Latitude vs. Temperature Plot

```python
# This scatter plot was created using Latitude for the x axis, and the Max Temperature for y. 
# We can see, as the latitude approaches 0, the temperature gets higher, however,
# the true peak for temperatures is more around 20 degrees latitude, than 0.
```

```python
x_axis = clean_city_data["Lat"]
max_temp = clean_city_data["Max Temp_C"]
plt.figure(figsize=(10,7))
plt.scatter(x_axis,max_temp, facecolor="red", edgecolor="black", linewidth=2)
plt.grid()
plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
```

## Latitude vs. Humidity Plot

```python
# In this scatter plot, we don't see much of a correlation between latitude and himidity.
# More countries surveyed overall have high humidity, and very few having low humidity.
```

```python
humidity = clean_city_data["Humidity"]
plt.figure(figsize=(10,7))
plt.scatter(x_axis,humidity, facecolor="red", edgecolor="black", linewidths=2)
plt.grid()
plt.title("City Latitude vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
```

## Latitude vs. Cloudiness Plot

```python
# ONce again, we do not see any strict correlation between latitude and cloudiness.
# We also see a much more even distribution of cloudiness %, than the previous plot, 
# telling us that cloudiness, humidity and temperature do not have a necesary correlation either.
```

```python
cloudiness = clean_city_data["Cloudiness"]
plt.figure(figsize=(10,7))
plt.scatter(x_axis,cloudiness, facecolor="red", edgecolor="black", linewidths=2)
plt.grid()
plt.title("City Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
```

## Latitude vs. Wind Speed Plot

```python
# Here we see another scatter plot that has no obvious correlation with the others.
# Most countries surveyed have wind speeds under 20MPH. However, we can see outliers both above and below 0 degrees latitude
# There is a dip in countries exceeding 20MPH wind speeds around 0 degrees latitude.
```

```python
wind_speed = clean_city_data["Wind Speed"]
plt.figure(figsize=(10,7))
plt.scatter(x_axis,wind_speed, facecolor="red", edgecolor="black", linewidths=2)
plt.grid()
plt.title("City Latitude vs. Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (MPH)")
```

## Linear Regression

```python
import scipy.stats as st

x_axis_nh = clean_city_data["Lat"].loc[clean_city_data["Lat"]>=0]
x_axis_sh = clean_city_data["Lat"].loc[clean_city_data["Lat"]<0]
max_temp_nh = clean_city_data["Max Temp_C"].loc[clean_city_data["Lat"]>=0]
max_temp_sh = clean_city_data["Max Temp_C"].loc[clean_city_data["Lat"]<0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
slope, intercept, r_value, p_value, std_err  = st.linregress(x_axis_nh, max_temp_nh)
plt.figure(figsize=(10,7))
plt.scatter(x_axis_nh,max_temp_nh, edgecolor="black")
plt.plot(x_axis_nh,(slope*x_axis_nh)+intercept, color="red", label="y={:.2f}x+{:.2f}".format(slope,intercept))
plt.legend(fontsize=20)
plt.title("Northern Hemisphere Max Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
print(f"The r value = {r_value}")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
slope, intercept, r_value, p_value, std_err  = st.linregress(x_axis_sh, max_temp_sh)
plt.figure(figsize=(10,7))
plt.scatter(x_axis_sh,max_temp_sh, edgecolor="black")
plt.plot(x_axis_sh,(slope*x_axis_sh)+intercept, color="red", label="y={:.2f}x+{:.2f}".format(slope,intercept))
plt.legend(fontsize=20)
plt.title("Southern Hemisphere Max Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
print(f"The r value = {r_value}")
```
